## ðŸ”·What is Generator?
A `generator` is a `special function that returns an iterator` and `produces value one at a time` using `yield` instead of `return`.

ðŸ“ŒSimple Definition

_A generator generates values one by one, only when needed._

## ðŸ”·Why Generators Are Needed (Problem -> Solution)

### âŒ Without Generator(List)

```python
def get_numbers():
    return [1,2,3,4,5]
get_numbers()


# Note: All values stored in memory at once.
```

## âœ… With Generator


```python
def get_numbers():
    for i in range(1,6):
        yield i

x= get_numbers()
print(next(x))


# Notes:
# Values created on demand
# Memory efficient
```

## ðŸ”· `yield` vs `return`

| `yield` | `return`|
|--|--|
|Pauses function|Ends function|
|Retuns multiple values| Return one values|
|Memory efficient|Memory heavy|

## ðŸ”· Simple Generator Example

```python
def simple_gen():
    yield 1
    yield 2
    yield 3

g = simple_gen()
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

## ðŸ”· Generator with Loop

```python
def count_up(n):
    for i in range(1,n+1):
        yield i
for num in count_up(5):
    print(num)
```

### ðŸ”· How Generator Works 

ðŸ“Œ steps:
1. Function is called --> no exection
2. Returns generator object
3. `next()` resumes execution
4. Stiops at `yield`
5. Continues from same point

### ðŸ”· Generator VS Normal function

|`Feature`|`Generator`|`Function`|
|--|--|--|
|Uses|`yield`|`return`|
|Memory|Low|High|
|Execution|Lazy|Immediate|
|values|One at a time|All at once|

## ðŸ”· Generator Expression (Like List comprehension)

### List Comprehension

```python
nums = [x*x for x in range(5)]
nums
```

### Generator Expression

```python
nums = (x*x for x in range(5))
print(next(nums))
print(next(nums))
print(next(nums))
print(next(nums))
print(next(nums))
```

### Example: Even Numbers Generator

```python
def even_Numbers(n):
    for i in range(n):
        if i%2 == 0:
            yield i


evenNum = even_Numbers(10)
print(next(evenNum))
print(next(evenNum))
print(next(evenNum))
print(next(evenNum))
print(next(evenNum))
# print(next(evenNum))
```

### Reading Large File using Generator

```python
def read_file(filename):
    with open(filename) as file:
        for line in file:
            yield line.strip()

x = read_file("C:\\Users\\User\\Desktop\\NursaryRhymes.txt")
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

## ðŸ”· Generator with `send()`

```python
def gen():
    value = yield 1
    yield value

g = gen()
print(next(g))
print(g.send(10))
```
