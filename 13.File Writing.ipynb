## File Writing Modes
|Mode|Meaning|
|----|---|
|w|Write(create new/overwrites file)|
|a|Append(adds data at end of the line)|
|w+|Write+Read|
|a+|Append + Read|


## Step-by-Step Process
* __1.Open the file.__
* __2.Write data__
* __3.Close the file.__

# 1. Writing using write()

```python
file = open("C:\\Users\\User\\Desktop\\data.txt","w")
file.write("Hello Python\n")
file.write("File Writing Example")
file.close()

# W mode Overwrites existing content 
# use \n for new line 
```

# 2. Writing Using writelines() - Multiple Lines

```python
lines = ["Python\n","Java\n","C++\n", "Go\n"]

file = open("C:\\Users\\User\\Desktop\\languages.txt","w")
file.writelines(lines)
file.close()

# Does not add newline automatically 
```

# 3. Append Mode(a) - Add Data without Deleting 

```python
file  = open("C:\\Users\\User\\Desktop\\languages.txt",'a')
file.write("\nNew Line added")
file.close()
```

# 4. Writing Files Using With

```python
with open("C:\\Users\\User\\Desktop\\info.txt","w")as file:
    file.write("Uisng with statement\n")
    file.write("File auto closes")


# Recommended approach
```

# 5. Writing User input to File 



```python
name = input("Enter your name: ")
age = input("Enter your age: ")


with open("C:\\Users\\User\\Desktop\\languages.txt","w") as file:
    file.write(f"Name: {name}\n")
    file.write(f"Age: {age}")
    
```

# 6. Writing Numbers to file



```python
numbers = [10,20,30]

with open("C:\\Users\\User\\Desktop\\languages.txt",'w') as file:
    for num in numbers:
        file.write(str(numbers)+"\n")
```

# 7. Write+Read(w+ Mode)

```python
with open('C:\\Users\\User\\Desktop\\languages.txt','w+') as file:
    file.write('Hello world')
    file.seek(0)
    print(file.read())
```

# 8. Writing Uisng Loop (Log file example)

```python
with open('C:\\Users\\User\\Desktop\\languages.txt',"w") as file:
    for i in range(3):
        file.write(f"Log entry:{i}\n")
```
