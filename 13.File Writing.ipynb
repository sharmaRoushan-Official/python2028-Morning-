   * __üìåUsed For__

## File Writing Modes
|Mode|Meaning|
|----|---|
|w|Write(create new/overwrites file)|
|a|Append(adds data at end of the line)|
|w+|Write+Read|
|a+|Append + Read|


## Step-by-Step Process
* __1.Open the file.__
* __2.Write data__
* __3.Close the file.__

# 1. Writing using write()

```python
file = open("C:\\Users\\User\\Desktop\\data.txt","w")
file.write("Hello Python\n")
file.write("File Writing Example")
file.close()

# W mode Overwrites existing content 
# use \n for new line 
```

# 2. Writing Using writelines() - Multiple Lines

```python
lines = ["Python\n","Java\n","C++\n", "Go\n"]

file = open("C:\\Users\\User\\Desktop\\languages.txt","w")
file.writelines(lines)
file.close()

# Does not add newline automatically 
```

# 3. Append Mode(a) - Add Data without Deleting 

```python
file  = open("C:\\Users\\User\\Desktop\\languages.txt",'a')
file.write("\nNew Line added")
file.close()
```

# 4. Writing Files Using With

```python
with open("C:\\Users\\User\\Desktop\\info.txt","w")as file:
    file.write("Uisng with statement\n")
    file.write("File auto closes")


# Recommended approach
```

# 5. Writing User input to File 



```python
name = input("Enter your name: ")
age = input("Enter your age: ")


with open("C:\\Users\\User\\Desktop\\languages.txt","w") as file:
    file.write(f"Name: {name}\n")
    file.write(f"Age: {age}")
    
```

# 6. Writing Numbers to file



```python
numbers = [10,20,30]

with open("C:\\Users\\User\\Desktop\\languages.txt",'w') as file:
    for num in numbers:
        file.write(str(numbers)+"\n")
```

# 7. Write+Read(w+ Mode)

```python
with open('C:\\Users\\User\\Desktop\\languages.txt','w+') as file:
    file.write('Hello world')
    file.seek(0)
    print(file.read())
```

# 8. Writing Uisng Loop (Log file example)

```python
with open('C:\\Users\\User\\Desktop\\languages.txt',"w") as file:
    for i in range(3):
        file.write(f"Log entry:{i}\n")
```

# 9. Exception Handling While Writing 

```python
try:
    with open("C:\\Users\\User\\Desktop\\languages.txt","w") as file:
        file.write("Safe Writing")
except IOE:
    print("File error occurred.")
except Exception as e:
    print("Error:",e)
```

### Summary Table 


|Mehtod|Use|
|---|----|
|write()|Write single string|
|writelines()|Write List of strings|
|w|Overwrite|
|a|Append|
|with|Safe & clean|

```python
# Example: Read from one and Write another

with open("C:\\Users\\User\\Desktop\\languages.txt",'w') as outfile:
    while True:
        val = input("Enter your text here: ")
        if val == "0":
            break
        
        outfile.write("languages List:")
        outfile.write(val)
        outfile.write("\n")
        
with open("C:\\Users\\User\\Desktop\\languages.txt",'r') as infile:
    print("----------------Content------------------")
    content = infile.read()
    print(content)
```

#  CSV File Writing using Python

### What is CSV?

 * __Comma Separated Values File__
 * __Used in Excel, databases, reports__

### ‚úÖ Required Library

```python
import csv
```

```python
# help(csv)
```

```python
import csv

with open("C:\\Users\\User\\Desktop\\students.csv",'w',newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["ID","Name","marks"])
    writer.writerow([101,"Rahul",85])
    writer.writerow([102,"Ankita",90])
    writer.writerow([103,"Kiran",78])
```

### Writing Excel File

```python
from openpyxl import Workbook
```

```python
wb = Workbook()
sheet = wb.active

sheet['A1'] = "ID"
sheet["B1"] = "Name"
sheet["C1"] = "Salary"

sheet.append([101,"Rakul Preet",40000])
sheet.append([102,"Ankita",45000])
sheet.append([103,"Mayuri",60000])

wb.save("C:\\Users\\User\\Desktop\\employee.xlsx")
```

## Mini Project - Student Report System

```python
name = input("Enter Student Name: ")
marks = int(input("Enter marks: "))


with open("C:\\Users\\User\\Desktop\\studentReport.txt",'w') as file:
    file.write("Student Report\n")
    file.write(f"Name: {name}\n")
    file.write(f"Marks: {marks}\n")

    if marks>=40:
        file.write(f"Result:PASS")
    else:
        file.write(f"Result:FAIL")
```
