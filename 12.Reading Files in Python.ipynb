# 12. File Handling 

 * __File handling allows Python programs to read data from files and write data to files permanently.__


### Why we read files:
 * __Store data permanently__
 * __Read logs, reports, CSVs__
 * __Process large data.__
 * __Avoid manual input__

### Types of Files
 * __1.Text files: (.txt,.csv,.log)__
 * __2.Binary files:(.jpg, .pdf, .exe)__

### Steps to Read a File in Python

 * __1. Open the file__
 * __2. Read the file__
 * __3. Close the file__


|Mode|Meaning|
|----------|----------|
|r|Read(default)|
|w|Write|
|a|Append|
|r+|Read & Write|

#### Method-1 read() -- Read Entire data from file

```python
file = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt",'r')
content = file.read()
print(content)
file.close()
```

### Method 2: readline() --> Read one Line from file




```python
file = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt",'r')
content = file.readline()
print(content)
file.close()
```

```python
### Method 3: readlines() --> Read all lines as list

file = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt",'r')
content = file.readlines()
print(content)
file.close()
```

### Method 4: Reading Using for Loop 

```python
file = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt","r")

for line in file:
    print(line,end=" ")

file.close()
```

### Best Way: with statmenet (Use always this method)

```python
with open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt", 'r') as file:
    content = file.read()
    print(content)
```

```python
# No need to close the file
# cleaner and safer
```

```python
with open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt","r") as file:
    for x in file:
        print(x.strip())
```

### Hndling File Not Found Error

```python
try:
    with open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt","r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found")
```

### Reading specific Number of chracters

```python
# Reads first 10 chracters from the file
with open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt",'r') as file:
    print(file.read(10))
```

### Checking current File Position 

```python
with open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt","r") as file:
    print(file.tell())  # index value where pointer belongs
    file.read(5)  # index value 5 will be returnted bcoz pointer set to index 5.
    print(file.tell())
    file.read(5)
    print(file.tell())
```

### Moving Cursor(seek())

```python
with open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt","r") as file:
    file.seek(0)
    print(file.read())
```

## 1. Reading PDF files in Python

#### What is a PDF?
 * __Protable Document Format__
 * __Used for reports, invoices, resumes.__
 * __Text is not directly readable --> needs a library.__


## Required Library
* __pip install PyPDF2__

```python
import PyPDF2
```

```python
# help(PyPDF2)
```

```python
file = open("C:\\Users\\User\\Desktop\\Twinkle.pdf",'rb')
reader = PyPDF2.PdfReader(file)
print("Number of Pages:",len(reader.pages))

page = reader.pages[0]
print(page.extract_text())

page1 = reader.pages[1]
print(page1.extract_text())
```

### Notes:
 * __Open PDF in binary mode (rb)__
 * __Some PDFs may not extract text(Scanned PDF)__
 * __for Scanned PDF -> OCR required(advanced topic)__

### What is XML?
 * __Extensible Markup Language.__
 * __Stores Structured data.__
 * __Used in APIs, configs, data Exchange.__

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse("C:\\Users\\User\\Desktop\\with-schema (1)\\with-schema.xml")
root  = tree.getroot()
print(root)

for data in root:
    print("----------------------------------------------------------------------")
    id = data.find('id').text
    firstname = data.find('firstName').text
    lstname = data.find('lastName').text
    position = data.find('position').text
    department = data.find('department').text
    hireDate = data.find('hireDate').text
    salary = data.find('salary').text
    print("id.","FirstName","LastName","position","Department","hireDate","Salary")
    
    print(id,".",firstname," ",lstname,",",position,",",department,",",hireDate,",",salary)
```

### Notes

 * __XMl is hierarchical (tree Structured)__
 * __root --> child --> subchild__
 * __Used in Web Services & configurations.__

|Feature|PDF|Word|XML|
