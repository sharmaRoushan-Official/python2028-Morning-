# 3.Lists in Python 

 * __List are ordered.__
 * __List can contains any type of data type.__
 * __List elemnets can be accessed by index.__
 * __List can be nested to any arbitrary depth.__
 * __Lists are mutable.__
 * __Lists are dynamic.__

```python
# Creating a list 

nlis = []
print(nlis,type(nlis))
```

```python
# create a list with elements

nlis = ["Python", 25,2025,True,False]
print(nlis,type(nlis))
```

```python
# Index value [Positive and Negative value]

# Positive indexing

print(nlis[0])
print(nlis[1])
print(nlis[2])
print(nlis[3])
print(nlis[4])
# print(nlis[5]) there is no value present on this address
```

```python
# Negative Indexing 

print(nlis[-1])
print(nlis[-2])
print(nlis[-3])
print(nlis[-4])
print(nlis[-5])
```

```python
# total length of the list 

print(len(nlis))
```

```python
nlis
```

```python
# slicing of list element  , n-1 = 

newList = nlis[0:3]  # [starts: ends]
print(newList)
```

```python
newList1 = nlis[3:5]  # n-1 = 4-1 = 3, 5-1 = 4
print(newList1)
```
