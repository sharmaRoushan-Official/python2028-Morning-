### len(), count(), index(), max(), min(), sum() function in list 

```python
lis = [1,2,3,4,5,6,7,8,9,1,2,2,2]  # all values are in number form. # aggrerate function
print(len(lis))  # 9

print(lis.count(2)) # How many 2 are on the list 'lis'?
print(lis.index(7)) # What is the index of the number 7 in the list/
print(max(lis)) # What is the maximum value in the list?
print(min(lis)) # it will return minimum value in the list
print(sum(lis)) # return total value after additing of the list
```

### Conversion of a string into a list using split() function 

```python
message = "Python is a programming language."
print(message,type(message))
y=message.split()
print(y, type(y))
```

### use of split() function with a delimiter


```python
txt = "p,y,t,h,o,n"
print(txt,type(txt))

txt2 = txt.split(",")
print(txt2,type(txt2))
```

### copy() the list 

### shallow vs deep copy

```python
# 1) shallow copy

originalValue = [[1,2],[3,4]]  # nested List
print(originalValue,type(originalValue))

copiedValue = originalValue.copy()
print(copiedValue,type(copiedValue))


originalValue.append("Python")
print(originalValue)
```

```python
copiedValue
```

```python
# Deep Copy()


deepCopy = originalValue
print(deepCopy,type(deepCopy))
```

```python
copiedDeepValue = deepCopy
print(copiedDeepValue)
```

```python
deepCopy.append("Hello world")
```

```python
deepCopy
```

```python
copiedDeepValue
```

```python
# eval() function 

# This function serves the aim of converting a string to an integer or a float


expression = "8+7"  # 
print(expression,type(expression))

total = eval(expression)
print(total,type(total))  # 15, integer
```
