### 4.API URL(EndPoint)

### 5. GET Request(Read Data)

```python
resp = requests.get(url)
resp
```

### Notes

* `requests.get` -> Sends GET request
* `resp` --> resp is a variable that storest server reply

```python
print(resp.status_code)
```

### Important Code Port

* `200` --> Sucess
* `404` --> API not found
* `500` --> Server Error

## Convert API Response to Python Format

# 8. POST Request (SEND DATA)

Create Data

```python
payload = {
    "name":"Kushal",
    "email":"kushal@gmail.com",
    "age":30,
    "Id":1,
}
```

```python
print(payload,type(payload))
```

## Notes
* payload = data we send

## Send POST Request


```python
response = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=payload
)
```

## Notes

* `post()` -> create data
* `json=payload` -> converts dict --> JSON

# View Response

```python
print(response.status_code)
```

```python
print(response.json)
```

```python
url = "https://jsonplaceholder.typicode.com/users"
print(url)
```

```python
data  = resp.json()
print(data)
```

## 9. PUT Request (UPDATE DATA)

```python
url = "https://jsonplaceholder.typicode.com/posts/1"
print(url)

updated_data = {
    "name" : "Update Kushal",
    "email":"updateKushal@gmail.com"
}

response = requests.put(url, json=updated_data)
print(response.json)
print(response.text)
```

```python
print(response.json)
```

## Notes:

PUT = update exisiting data

## 10.DELETE REQUEST (Remove Data)



```python
url = "https://jsonplaceholder.typicode.com/posts/1"

response = requests.delete(url)
print(response.status_code)
```

## Expected Output:

`200` or `204`

```python
print(response.json)
print(response.text)
```

## 11. API Headers (Basic Authentication Concept)

```python
headers={
    "Authorization":"Bearer TOKEN_VALUE",
    "Content-Type":"application/json"
}
response = requests.get(url,headers=headers)
print(response)
```
