# 16. Decorators in Python

 * __What is a Decorator?__
 * A `decorator` is a __function that modifies another function's behavior without changing its code.__

## Simple Difinition
 * üìå _A decorator adds extra features to a function._

## üî∑ Why Decorators Are Needed

__‚ùå Without Decorators__
If you want to: 
 * __Log function calls__
 * __Check User authentication__
 * __Measure execution time.__
   
You would have to repeat the same code in every function:

__‚úÖWith Decorator__
 * __Write the logic once__
 * __Reuse it for many functions.__
 * __Cleaner & maintable code.__

## üî∑ Real-Life Example

üéÅ Gift Wrapper
* Gift = Original Function
* Wrapper = Decorator
* Gift remain same, wrapper adds value

## üî∑ Basic Decorator Structure

```python
def decorator(func): # func = say_hello
    def wrapper():
        print("Before Function")
        func()  # say_hello()  # calling function
        print("After Function")
    return wrapper
```

## üî∑ Applying a Decorator(Manual)

```python
def say_hello():
    print("Hello world")


say_hello = decorator(say_hello)
say_hello()
```

```python
# Ex-2


def sum_two():
    a = 10
    b= 15
    print(f"The sum of Two {a} and {b}:{a+b}")

# calling this function with decorator


func2 = decorator(sum_two)
func2()
```

## üî∑ Using @ Syntax(Recommended)

```python
@decorator
def say_hello():
    print("Congratulation World")

say_hello()
```

## üî∑ Decorator with arguments

```python
def decorator(func):  # func = greet
    def wrapper(name):
        nameUpper = name.upper()
        print("Welcome!")
        func(nameUpper)
    return wrapper

@decorator
def greet(name):
    print("Hello",name)
greet("gaurav")
```

## üî∑ Decorator with *args and ** kwargs

```python
def decorator(func):
    def wrapper(*args,**kwargs):
        print("Function started")
        result = func(*args,**kwargs)
        print("Function ended")
        return result
    return wrapper
```

```python
import time

# help(time)

def timer(func):
    def wrapper(*args,**kwargs):
        start = time.time()
        print("Starts at:",start)
        func(*args, **kwargs)
        end = time.time()
        print("Ends at:",end)
        print("Execution Time:", end - start)
    return wrapper


@timer
def demo():
    for i in range(1000000000):
        pass
demo() 
```
