# 10. Exception Handling 

 * __An exception is an event, which occurs during the execution of a program that disrupts the normal flow of the program's instrcutions.__
 * __In general, when a Python script encounters a situation that it cannot cope with, it raises exception.__
 * __An exception is a python object that represents an Error.__
 * __When a  Python script raises an exception, It must either handle the exception immediately otherwise it  terminates and quits.__
 * __If you have some suspicious code that may raise an exception, you can defend your program by placing the suspicious code in a try:block.__
 * __After the try:block, include an except:statement, followed by a block of code which handles the problem as elegantly as possible.__
 * __Common Excetpions.__
   * __ZeroDivisionError__
   * __ValueError__
   * __IOError__
   * __IdentationError__

```python
# 4. else Block

 # Purpose:
    # Exceutes only if no exception Occurs.
 # KeyPoint
    # Runs after try if everything is successful.
    # Helps separate error-prone code from success logic
    # Improve readability.

# Tips:
# Use else for bussiness logic that should run only when try succeed.
```

```python
try:
    num = int(input("Enter any value: "))

except ValueError:
    print("Invalid Input")
else:
    print("You Entered value:",num)
```

```python
# 5.finally Block
 # Purpose:
    # Executes always, whether an exception occurs or not.
 # KeyPoint
    # Used for cleanup operations
    # Executes even if return, break, or exception occurs
    # Common uses:
    # closing Files.
    # Closing database conncetions
    # Releasing resourses.
```

```python
try:
    f = open("C:\\Users\\User\\Desktop\\NursaryRhymes.txt",'r')
    print(f.read())
except FileNotFoundError:
    print("File not Found")
finally:
    print("Closing the file")
    f.close()
```

```python
#10.Write a program that tries to read from a file and ensures the file is closed using finally.

file_handle = None
try:
    file_handle = open('data.txt', 'r')
    content = file_handle.read()
    print("File contents:", content)
except FileNotFoundError:
    print("Error: data.txt not found.")
finally:
    if file_handle is not None:
        file_handle.close()
        print("File closed safely.")
```

```python
#11.Write a program that raises a custom exception if the user enters a negative number.

class NegativeNumberError(ValueError):
    pass

try:
    user_input = input("Enter a positive number: ")
    number = int(user_input)
    if number < 0:
        raise NegativeNumberError("Negative numbers not allowed! try again.")
    print("You entered:", number)
except ValueError as e:
    if isinstance(e, NegativeNumberError):
        print(f"Custom error: {e}")
    else:
        print("Please enter a valid integer.")
```

```python
#12.Use try–except–else–finally to validate user login (username and password check).
VALID_USER = "admin"
VALID_PASS = "2026"

try:
    username = input("Enter username: ")
    password = input("Enter password: ")
    if not username or not password:
        raise ValueError("Username and password cannot be empty.")
except ValueError as e:
    print(f"Input error: {e}")
else:
    if username == VALID_USER and password == VALID_PASS:
        print("Login successful! Welcome,", username)
    else:
        print("Invalid credentials.")
finally:
    print("Login process completed.")
```
