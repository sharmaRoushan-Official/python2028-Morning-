# 6. Dicationary in Python

 * __Dicationaries are used to store data in key:value pairs.__
 * __A dictionary is a collection which is ordered, changeable or mutable and do not allow duplicates.__
 * __Dicationary items are presented in key:value pairs, and can be referred to by using the key name.__
 * __Dicationaries cannot have two items with same key.__
 * __A dictionary can nested and can contain another dictionary.__


 * __Keys ----------- Values__
   * __'name' : "Bob"__
   * __'age'  : 30__
   * __'job'  : "DevOps"__
   * __'city' : "Texas"__
   * __'email' : "bob@hotmail.com"__

```python
x = {}  # set, dictionary
print(x,type(x))
```

```python
x = {1}  # set, dictionary
print(x,type(x))
```

```python
x1 = set()
print(x1,type(x1))
```

```python
# take a sample dictionary 
sample_dict = {'key_1':3.147, 'key_2':1.618,
              'key_3':True, 'key_4':[3.147,1.618],
              'key_5':(147,100,123.22),'key_6':2025}
print(sample_dict)

```

## Note: As you see that the whole dicationary is enclosed in curly brackets, each key is seperated from its value by a colon ":" and , commas are used to seperated the items in the dicationary.

```python
# fromkeys() function 

# it returns a new dicationary with the certain sequence of the items as the keys of the dicatonary and values are assigned with None. 

keys = {'A','T','C','G'}  # it is a set
print(keys, type(keys))

# set to dicationary conversion 
sequence = dict.fromkeys(keys)
print(sequence,type(sequence))
```
