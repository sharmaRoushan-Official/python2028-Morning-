# String Operations 


```python
message = "hello PYThon"
print("Original String:",message)
print(type(message))


# upper case

msgUpper = message.upper()
print(msgUpper)

# lower case

msgLower = message.lower()
print(msgLower)


# capitalize case
msgCapitalize = message.capitalize()
print(msgCapitalize)

# title case
msgTitle = message.title()
print(msgTitle)
```

```python
# replace() 

message = "Hello Python Python Python"
print(message)

message2 = message.replace("Python","World")
print(message2)
```

```python
# find()

message = "Hello world"

print(message.find("world"))
```

```python
print(message.find("w"))
```

```python
print(message.find("apple")) 
```

```python
print(message.find("lo"))
print(message.find("ld"))
```

```python
# formatting of string

message = "Hello {Python} Python Python"  # {place hold}
print(message)

print(message.format(Python="World"))
```

```python
print(message.format(Python="Kushal"))
```

```python
message1 = "hi,MY name is {} and {} years old"
print(message1.format("Kushal",35))
```

```python
message1 = "hi,MY name is {} and {} years old"
print(message1.format(35,66))
```

```python
message1 = "hi,MY name is {a} and {b} years old"
print(message1.format(a="Kushal",b=66))
```

```python
message1 = "hi,MY name is {a} and {b} years old"
print(message1.format(b=35,a="Kushal"))
```
