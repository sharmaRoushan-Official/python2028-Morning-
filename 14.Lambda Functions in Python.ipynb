# 14. Lambda Functions in Python

### What is Lambda Function?
 * __A Lambda function is a small, anonymous (nameless) function written in one single line.__


## ðŸ“ŒSimple Definition
 * __A lambda function is a one-line function without a name.__


## Why Lambda Functions are Used
 * __Short & Clean code.__
 * __Used for temporary operations.__
 * __Common with map(), filter(), reduce()__
 * __Improves readability in simple logic.__

## Normal Function vs Lambda function

```python
# simple function
def square(x):
    return x*x
square(10)
```

```python
# lambda function
(lambda x:x*x)(10)
```

### Notes
 * __Same Output__
 * __Less code__


### Syntax of Lambda function

```python
lambda arguments:expression
```

## ðŸ“Œ Rules:
 * __Can have multiple arguments__
 * __Only one expression__
 * __Automatically returns result__
 * __No return keyword__

## 1ï¸âƒ£ Addition of Two Numbers 

```python
# Normal Function
def addTwo(x,y):
    return x+y
addTwo(10,-50)
```

```python
(lambda a,b:a+b)(10,-50)
```

## 2ï¸âƒ£ Square of a number

```python
# Normal function

def sqrNum(x):
    return x**2

sqrNum(10)
```

```python
(lambda x:x**2)(10)
```

## 3ï¸âƒ£ Check Even or Odd Numbers

```python
def checkEvenorOdd(x):
    if x%2==0:
        return f"Even Number:{x}"

checkEvenorOdd(10)
```

```python
(lambda a:f'Even Number:{a}' if a%2==0 else f'Odd Number:{a}')(11)
```

## Lambda with Multiple Arguments


```python
# Normal Function
def maxNum(a,b,c):
    if a>b and a>c:
        print(f'Max Num:{a}')
    else:
        if b>c:
            print(f"Max Num: {b}")
        else:
            print(f"Max Num:{c}")
maxNum(10,20,15)
```

```python
# Anonymous Function
(lambda a,b,c: f"Max Num:{a}" if a>b and a>c else (f"Max Num:{b}" if b>c  else f"Max Num:{c}"))(10,20,15)
```

## Lambda with map() [Very Important]
 * __Apply a function to each element in a list__

```python
numbers = [1,2,3,4]

# syntax: map(function,sequence)

# with normal function 
def sqr(x):
    return x*x
```

```python
list(map(sqr,numbers))
```

```python
# syntax: map(function,sequence)

# with anonymous function
list(map(lambda x:x*x,numbers))
```
