# 9. Functions in Python

 * __In python, a function is a group of related statements that performs a specific task.__
 * __Functions helps to break our program into smaller and modular chunks. As per our program grows larger and larger, functions make it more organized and manageable.__
 * __Furthermore, it avoids repetation and makes the code reusable.__
 * __There are two types of functions:__
   * __Predefined Functions__
   * __Userdefined functions.__
 * __In Python a function is defined using the def keyword followed by the funciton name and parentheses().__
 * __Keyword def that marks to start of the function header.__
 * __A function name to uniquely identify the function.__
 * __Function naming follows the same rules of writing identifiers in Python.__
 * __Parameters(arguments) through which we pass values to a function. They are optional.__
 * __A colon(:) to mark the end of the function header.__
 * __Optional documentation string(docstring) to describe what the function does.__
 * __One or more valid python statements that mark up the function body.__
 * __Statements must have the same indentation level (usually 4 spaces).__
 * __An optinal return statement to return a value from the function.__

```python
def sum1(x):  # x is parameter
    y1 = x-8 # 10-8
    y2 = x+8 # 10+8
    y3 = x*8 # 10*8
    y4 = x/8 # 10/8
    y5 = x//8 # 10//8
    y6 = x%8  # 10%8
    print(f"If you make the above operations with {x}, the result will be {y1},{y2},{y3},{y4},{y5},{y6}")
    return y1,y2,y3,y4,y5,y6

sum1(10) # function invoking : 10 is argument : function call
```

```python
#3.Write a function to add two numbers.

def add_numbers(a, b):
    return a + b


result1 = add_numbers(5, 3)
result2 = add_numbers(10.5, 2.7)
result3 = add_numbers(-2, 7)

#test cases
print(f"5 + 3 = {result1}")
print(f"10.5 + 2.7 = {result2}")
print(f"-2 + 7 = {result3}")
```

```python
#4.Write a function to find the *square* of a number.

def square(number):
    return number * number

#test cases
print(f"Square of 5 = {square(5)}")
print(f"Square of 3 = {square(3)}")
print(f"Square of 4 = {square(4)}")
print(f"Square of 9 = {square(9)}")
```

```python
#5.Write a function to check whether a number is *even or odd*.

def is_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Test cases
print(f"45 is {is_even_odd(45)}")
print(f"60 is {is_even_odd(60)}")
print(f"0 is {is_even_odd(0)}")
print(f"3 is {is_even_odd(3)}")
print(f"2 is {is_even_odd(2)}")
```

```python
#6.Write a function to find the *maximum of two numbers*.

def find_max(a, b):
    if a > b:
        return a
    else:
        return b

# Test cases
print(f"Max(30, 5) = {find_max(30, 5)}")
print(f"Max(3.7, 3.8) = {find_max(3.7, 3.8)}")
print(f"Max(-5, -10) = {find_max(-5, -10)}")
print(f"Max(7, 7) = {find_max(7, 7)}")
```

```python
#7.Write a function to calculate the *area of a rectangle*.

def rectangle_area(length, width):  #area = length * width
    return length * width

# Test cases
print(f"Area(5, 3) = {rectangle_area(5, 3)}")
print(f"Area(4.5, 2.8) = {rectangle_area(4.5, 2.8)}")
print(f"Area(10, 2) = {rectangle_area(10, 2)}")
print(f"Area(6.2, 3.1) = {rectangle_area(6.2, 3.1)}")
```

```python
#9.Write a function to check if a number is *positive, negative, or zero*.

def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"
    
#test cases
print(f"10 is {check_number(10)}")
print(f"-5 is {check_number(-5)}")
print(f"0 is {check_number(0)}")
```

```python
#10.Write a function to return the *length of a string*.

def string_length(text):
    count = 0
    for char in text:
        count += 1
    return f'Length of {text} = {count}'
    
#test cases
print(string_length("Hello"))
print(string_length("this"))
print(string_length("is"))
print(string_length("Python"))
print(string_length("Lecture"))
```
