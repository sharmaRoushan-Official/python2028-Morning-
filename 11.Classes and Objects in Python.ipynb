```python
# find the area of circle,square using class and object radius value is 10cm and side is 8cm

class circle:
    def __init__(self,radius):
        self.radius = radius


    def areaCircle(self):
        area = 3.147*self.radius * self.radius
        return f"The Area of Circle is: {area} cm²"

obj1 = circle(10)
obj1.areaCircle()
```

```python
class square:
    def __init__(self, side):
        self.side = side

    def areaSquare(self):
        area = self.side * self.side
        print(f"Area of Square: {area} cm²")
    
obj2 = square(10)
obj2.areaSquare()
```

### Meaning of Self parameter:

 * __Self represnts the current object(instance) of a class.__
 * __It allows access to instance variable and instance methods.__
 * __It refers to "the object calling the method"__
 * __Without self, Python cannot know which object's data to use.__


### self is Not a Keyword
 * __Self is not a reversed keyword__
 * __It is convention.__
 * __You can rename it(i.e, this), but strongly discouraged.__

### self is Passed Automatically
 * __When a method is called using an object.__
 * __obj.method()__
 * __python internally converts it to:__
 * __obj is received as self__

### self vs Class Variable 

```python
class Demo:
    count = 0  # class variable

    def __init__(self):  # constructor
        self.count = 10  # Instance variable

    def displayIns(self):
        return f"{self.count}"

obj = Demo()
print(obj.count)  # Class Variable

print(obj.displayIns()) # Instance Varibale
```

```python
# 1. Find the area of triangle and their perimeter, side is 45,55,88
class Triangle:
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        # Rule:
        # if you sum to two sides of triangle then the value will be greater than 3rd one.

    def area(self):
        s = (self.side1 + self.side2 + self.side3) / 2
        area = (s * (s - self.side1) * (s - self.side2) * (s - self.side3)) ** 0.5
        return area

    def perimeter(self):
        return self.side1 + self.side2 + self.side3

obj = Triangle(45,55,88)
obj.area()
```

```python
class parallelogram:
    def __init__(self, side1, side2):
        self.side1 = side1
        self.side2 = side2
        

    def area(self):
        area = self.side1 *  self.side2 
        return area,'cm²'

    def perimeter(self):
        peri = (self.side1*2) + (self.side2*2)
        return peri,"cm"
    
obj1=parallelogram(20,10)
print(obj1.area())
print(obj1.perimeter())
```
