```python
# 1. Find the area of triangle and their perimeter, side is 45,55,88
class Triangle:
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        # Rule:
        # if you sum to two sides of triangle then the value will be greater than 3rd one.

    def area(self):
        s = (self.side1 + self.side2 + self.side3) / 2
        area = (s * (s - self.side1) * (s - self.side2) * (s - self.side3)) ** 0.5
        return area

    def perimeter(self):
        return self.side1 + self.side2 + self.side3

obj = Triangle(45,55,88)
obj.area()
```

```python
class parallelogram:
    def __init__(self, side1, side2):
        self.side1 = side1
        self.side2 = side2
        

    def area(self):
        area = self.side1 *  self.side2 
        return area,'cm²'

    def perimeter(self):
        peri = (self.side1*2) + (self.side2*2)
        return peri,"cm"
    
obj1=parallelogram(20,10)
print(obj1.area())
print(obj1.perimeter())
```

```python
# Hybird Inheritance 

# Base Class/ Parent
class a:
    def showA(self):
        return "Class A Method"

# Derived class (Hierarchical Inheritance) 
class b(a):
    def showB(self):
        return "Class B Method"

# Derived Class  (Hierarchical Inheritance)
class c(a):
    def showC(self):
        return "Class C Method"
# Derived Class (Multiple Inherance--> hybrid)
class d(b,c):
    def showD(self):
        return "Class D Method"
obj1 = d()
print(obj1.showA())
print(obj1.showB())
print(obj1.showC())
print(obj1.showD())
```

```python
# Base Class/parent
class Person:
    def person_info(self):
        return "I am a person"

# Derived class  [Hierarchical Inheritance]
class Employee(Person):
    def employee_info(self):
        return "I am an Employee"
        
# Derived Class  [Hierarchical Inheritance] 
class Trainer(Person):
    def trainer_info(self):
        return "I am a trainer"

class TeamLead(Employee,Trainer): # Multiple+Hierarchical
    def teamlead_info(self):
        return "I am a team lead"
t1 = TeamLead()
print(t1.person_info())
print(t1.employee_info())
print(t1.trainer_info())
print(t1.teamlead_info())
```

```python
# 1. Method Overloading (Python Style)


# Python does not support traditional Overloading, but we achieve it using default arguments.
```

```python
class Calculator:
    def add(self,a,b=0, c=0):
        return a+b+c

calc = Calculator()
print(calc.add(10))
print(calc.add(10,20))
print(calc.add(10,20,30))


# Same method name add
# Different number of arguments
# Behavior changes automatically.
```

```python
# 2. Method Overriding 

# Child class provides its own implementation of parent method 


# Example 
```

```python
class Employee:
    def role(self):
        return "I am an Employee"


class Trainer(Employee):
    def role(self):
        return "I am a Trainer."

class TeamLead(Employee):
    
    def role(self):
        return "I am a Team Lead."


e = Employee()
print(e.role())
t = Trainer()
print(t.role())

t1 = TeamLead()
print(t1.role())


# Same method role()
# Different Output
# Based on Object 
# Runtime Polymorphism
```

# Encapsulation

 * __Encapsulation means wrapping data(variables) and methods(functions) together into a single unit(class) and restricting direct access to data.__

 * __Simple Definition__

   * __Encapsulation = Data Hinding + Data protection.__
  
### Why Encapsulation is Needed 
 * __Without Encapsulation.__
     * __Anyone can change data.__
     * __No control on values or varibale.__
     * __Leads to bugs__
     * __No Security.__
    
 * __With Encpsulation__
     * __Data is Protected.__
     * __Controlled access.__
     * __Improve Security__
     * __Improves maintainability.__


### Encapsulation in Python

 * __Encapsulation is achieved using Access Modifiers.__


| Access Modifier | Syntax | Meaning |
|----------|----------|----------|
|Public|variable| Accessible everywhere|
|Protected|_varible| Accessible in class & child|
|Private| __varible| Accessbile only inside class|

```python
# 1. Public Members 


class Employee:
    def __init__(self,name): 
        self.name = name # public variblic

emp = Employee("Rahul")
print(emp.name)


# No restrictions
# Anyone can access this varible i.e self.name
```

```python
# Protected Members(_)


class Employee:
    def __init__(self,salary): # 10000
        self._salary = salary  # protected Varibales


    def showSalary(self):
        return f"Your Salary protected value is:{self._salary}"

emp = Employee(10000)
# emp.salary  # directed salary can't be accessed bcoz it is protected.
emp.showSalary()   
```

```python
# accessing Private Variable
class Manager(Employee):
    def show_salary(self):
        print(self._salary)
m = Manager(10000)
m.show_salary()
```

```python
# Private Members (_ _) - Most Important 

class BankAccount:
    def __init__(self):
        self.__balance = 10000 # private varibale

    def show_balance(self):
        print("Balance:",self.__balance)


    def withdraw(self):
        amount = int(input("Enter your amount here: "))
        if amount<= self.__balance:
            self.__balance -= amount
            print(f"Withdrawal Successful:₹ {amount}\nCurrent Balance is:₹{self.__balance}")
        else:
            print("Insufficient Funds!!")
            


bank = BankAccount()
# bank.__balance  # Not allowed: Error
bank.show_balance()
bank.withdraw()
```

```python
# Encapsulations Using Getter & Setter Methos 

class Student:
    def __init__(self):
        self.__marks =0

    # setter method to private varibale
    def set_marks(self):
        marks = int(input("Enter your marks: "))

        if marks>=0 and marks <=100:
            self.__marks = marks
        else:
            print("Invalid Marks")

    # getter method
    def get_marks(self):
        return f"You Scored: {self.__marks}"


s = Student()
s.set_marks()
```

### Encapsulation using @property Decorator (Advanced & Interviewer)

```python
class Employee:
    def __init__(self):
        self.__salary = 0

    @property
    def salary(self):
        return self.__salary  # value can't be return directly

    def salary(self):
        amount = int(input("Set Salary: "))

        if amount>0:
            self.__salary = amount
        else:
            print("Invalid Salary Value")

emp = Employee()
emp.salary()
print(emp.salary)
```

# 4. Data Abstraction


 * __Data Abstraction means hiding internal implementation details ans showing only essential features to the uers.__
 * __Notes: Whatn an object does is shown, how it does is hidden.__



## Why Data Abstraction is Needed
### Without Abstraction

 * __User sees unnecesary details.__
 * __Code becomes complex.__
 * __Difficult to maintain.__
 * __Higher chances of misuse.__

### With Abstraction
 * __Only required features exposed.__
 * __Internal Logic hidden.__
 * __Easy to use and Maintain.__
 * __Improves Security.__


## Ex:-
 * __Real-Life Example:__
   * __Driving a Car:__
   * __You Use:__
     * __Stering__
     * __Accelerator__
     * __Break__
   * __You don't know engine internals.__


### Data Abstraction VS Encapsulation

|Feature | Encapsulation | Abstraction |
|----------|----------|----------|
| Focus | Data hiding | Implementation Hinding| 
| Shows| how data is protected| What functionality is provided.|
|Achieved by| Access Modifier| Abstract Base Classes|Interfaces|


### Note: 
 * __Encapsulation: How Data is stored.__
 * __Abstraction: What user can do with data.__


### How Abstraction is Achieved in Python
 * __Python provides:__
 * __Abstract Base Classes(ABC)__
 * __Abstract Method__

```python
# Syntax for Abstraction in Python 

from abc import ABC, abstractmethod 
```

```python
help(ABC)
```

```python
from abc import ABC, abstractmethod 


class Vehicle(ABC):

    @abstractmethod
    def start(self):
        # return "Engine Start"
        pass
```

```python
# cannot create object of this class 

v = Vehicle()
v.start()
```

```python
# Child Class Implementation Abstract method 


class car(Vehicle):

    def start1(self):
        print("Car Start with Key")

    def start(self):
        pass


c = car()
c.start1()
```

```python
class Bike(Vehicle):
    
    def break1(self):
        print("Bike Break Applied")

    def start(self):
        return "Bike Engine Starts"


b = Bike()
b.break1()
b.start()
```

# Important Points

#### Vehicle defines what should be done 
#### Car and Bike define how it done 
#### User only calls start()
